### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

 ### Answer: In Pandas, a DataFrame is a two-dimensional labeled data structure with columns of potentially different data types. It is similar to a spreadsheet  or a  SQL table. In a DataFrame , data is organized in rows and columns, where each column can be considered as a Pandas Series.

###         On the other hand, a Series in Pandas is a one-dimensional labeled array that can contain data of any data type, including integers, strings, and floating-point numbers.
       
###          Here's an example to illustrate the difference between a DataFrame and a Series:

###         Suppose we have the following data: Name(Anoop , Ravi, Rohit) , Age(25, 30, 32) , Gender (Male ,Male ,Male) We can represent this data using a DataFrame in Pandas as follows:

```python
data = {'Name': ['Anoop', 'Ravi', 'Rohit'],
        'Age': [25, 30, 32],
        'Gender': ['Male', 'Male', 'Male']}

df = pd.DataFrame(data)

print(df)
```

### Q.6 Which of the following is mutable in nature Series, DataFrame, Panel?


### Answer. Both Series and DataFrame are mutable in nature, while Panel is considered deprecated and has been replaced by multi-dimensional arrays, which are also mutable.

###     In a mutable object, the contents of the object can be modified after it is created. This means that you can add, remove, or update elements in a Series or DataFrame after they are created.
